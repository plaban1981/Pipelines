## Missing Values 

#### check the percentage of nan values present in each feature

*  make the list of features which has missing values
*  print the feature name and the percentage of missing values

```
features_with_nan = [feature for feature in df_train.columns if df_train[feature].isnull().sum() > 1]
```

```
for feature in features_with_nan:
  print(f"Feature Name : {feature} {np.round(df_train[feature].isnull().mean(),4)} % of missing value ")
```

#### Since they are many missing values, we need to find the relationship between missing values and Sales Price

```
data = df_train.copy()
for feature in features_with_nan:
  # let's make a variable that indicates 1 if the observation was missing or zero otherwise
  data['feature'] = np.where(data[feature].isnull(),1,0)
  # let's calculate the mean SalePrice where the information is missing or present
  data.groupby('feature')['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

Herewith  the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section

## Numerical Variables

```
# list of numerical variables
numerical_features = [feature for feature in df_train.columns if df_train[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
df_train[numerical_features].head()
```

## Temporal Variables(Eg: Datetime Variables)

From the Dataset we have 4 year variables. 

We have extract information from the datetime variables like no of years or no of days. 

One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold.

```
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

print(year_feature)
```

## let's explore the content of these year variables

```
for feature in year_feature:
    print(feature, df_train[feature].unique())
```

## Lets analyze the Temporal Datetime Variables
#### Check whether there is a relation between year the house is sold and the sales price

```
df_train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Medain Sale Price')
plt.title('House Price vs Year Sold')
plt.show()
```

here we see as the years go by the price decreases which does not possibly look ok

## Here we will compare the difference between All years feature with SalePrice and capture the difference between year variable and year the house was sold for

```
year_feature
```

```
for feature in year_feature:
  if feature != 'YrSold':
    data = df_train.copy()
    data[feature] = data['YrSold'] - data[feature] # difference between Year initiated and Year the house was sold
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```


## Numerical variables are usually of 2 types 
* Continous variable 
* Discrete Variables

Discrete variables will be having fixed set of integer values

```
#each and every category in the numerical varibale the threshold is considered to be 25, i.e. having 25 unique values
discrete_feature=[feature for feature in numerical_features if len(df_train[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```
discrete_feature
```

```
df_train[discrete_feature].head()
```

## Find the realtionship between them and Sale PRice

```
for feature in discrete_feature:
  data = df_train.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Median Sale Price')
  plt.title(feature + ' vs Sale Price')
  plt.show()
```

##Continuous Variable

```
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print(f"Continuous feature Count {len(continuous_feature)}")
```

```
continuous_feature
```

## In order to understand the distribution of Continuous variables we would use Histogram

```
for feature in continuous_feature:
  data = df_train.copy()
  data[feature].hist(bins=30)
  plt.xlabel(feature)
  plt.ylabel('Count')
  plt.title(feature)
  plt.show()
```

only SalePrice seems somewhat normally distributed.

Whenever the data is not normally distibuted or skewed we usually apply log Normal Distribution 

## Applying Logarithmic Transformation on skewed Continuous Variables 

One reason for this is to improve the linearity of the data.

np.log() will transform the variable, and np.exp() will reverse the transformation.

```
for feature in continuous_feature:
    data=df_train.copy() # storing a  separate copy , refereing to different memory location
    if 0 in data[feature].unique():#
        print(feature)
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

## Detecting Outliers- BoxPlot

Boxplot works only on continuous variables 

```
for feature in continuous_feature:
    data=df_train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
